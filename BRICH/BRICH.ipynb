```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets.samples_generator import make_blobs
from sklearn.cluster import Birch
from sklearn import metrics
from sklearn.grid_search import GridSearchCV
```

```python
X, Y=make_blobs(n_samples=1000, n_features=2, centers=[[-1, -1],[0,0],[1,1],[2,2]],
                cluster_std=[0.4, 0.3, 0.4, 0.3], random_state=9)
plt.scatter(X[:, 0],X[:, 1],marker='o')
plt.show()
```

```python
y_pred = Birch(n_clusters=None).fit_predict(X)
plt.scatter(X[:, 0],X[:, 1], c=y_pred)
plt.show()
print("Calinski-Harabaz Score", metrics.calinski_harabaz_score(X, y_pred))
```

```python
y_pred = Birch(n_clusters=4).fit_predict(X)
plt.scatter(X[:, 0],X[:, 1], c=y_pred)
plt.show()
print("Calinski-Harabaz Score", metrics.calinski_harabaz_score(X, y_pred))
```

```python
y_pred = Birch(n_clusters=4, threshold=0.2).fit_predict(X)
plt.scatter(X[:, 0],X[:, 1], c=y_pred)
plt.show()
print("Calinski-Harabaz Score", metrics.calinski_harabaz_score(X, y_pred))
```

```python
y_pred = Birch(n_clusters=4, threshold=0.3).fit_predict(X)
plt.scatter(X[:, 0],X[:, 1], c=y_pred)
plt.show()
print("Calinski-Harabaz Score", metrics.calinski_harabaz_score(X, y_pred))
```

```python
y_pred = Birch(n_clusters=4, threshold=0.3, branching_factor=20).fit_predict(X)
plt.scatter(X[:, 0],X[:, 1], c=y_pred)
plt.show()
print("Calinski-Harabaz Score", metrics.calinski_harabaz_score(X, y_pred))
```

```python
y_pred = Birch(n_clusters=4, threshold=0.3, branching_factor=30).fit_predict(X)
plt.scatter(X[:, 0],X[:, 1], c=y_pred)
plt.show()
print("Calinski-Harabaz Score", metrics.calinski_harabaz_score(X, y_pred))
```

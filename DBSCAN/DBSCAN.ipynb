```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets.samples_generator import make_circles, make_blobs
from sklearn import metrics
```

```python
# 生成数据
X1, y1 = make_circles(n_samples=5000, noise=0.05, random_state=9, factor=0.6)
X2, y2 = make_blobs(n_samples=1000, n_features=2, centers=[[1.2, 1.2]], cluster_std=[[.1]], random_state=9)
X = np.concatenate((X1,X2))
plt.scatter(X[:, 0], X[:, 1], marker='o')
plt.show()
```

```python
from sklearn.cluster import KMeans
# 先用kmeans尝试聚类
y_pred = KMeans(n_clusters=3, random_state=8).fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=y_pred, marker='o')
print(metrics.calinski_harabaz_score(X, y_pred))
```

```python
# 用DBSCSAN聚类

from sklearn.cluster import DBSCAN
y_pred = DBSCAN().fit_predict(X)
plt.scatter(X[:, 0], X[:, 1],c=y_pred, marker='o')
```

```python
y_pred = DBSCAN(eps=0.1, min_samples=10).fit_predict(X)
plt.scatter(X[:, 0], X[:, 1],c=y_pred, marker='o')
```
